```python
# 01_data_loading_and_initial_eda.ipynb

import pandas as pd
import numpy as np
import requests

# --------------------- Load Training Data ---------------------
url_train = "https://raw.githubusercontent.com/Bonface5/CUSTOMER-CHURN-PREDICTION-USING-ML/refs/heads/main/customer_churn_dataset-training-master.csv"

response = requests.get(url_train, stream=True)
with open("train.csv", "wb") as f:
    for chunk in response.iter_content(chunk_size=8192):
        if chunk:
            f.write(chunk)

train = pd.read_csv("train.csv")
train.head()

# --------------------- Load Test Data ---------------------
url_test = "https://raw.githubusercontent.com/Bonface5/CUSTOMER-CHURN-PREDICTION-USING-ML/refs/heads/main/customer_churn_dataset-testing-master.csv"

response = requests.get(url_test, stream=True)
with open("test.csv", "wb") as f:
    for chunk in response.iter_content(chunk_size=8192):
        if chunk:
            f.write(chunk)

test = pd.read_csv("test.csv")
test.head()

train.info()
train.isnull().sum()
train.describe(include="all")
```
